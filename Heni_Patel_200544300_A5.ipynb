```python
import pandas as pd

file_path = "C:/Users/patel/tripadvisor_hotel_reviews.csv"

data = pd.read_csv(file_path)

df = pd.DataFrame(data)

df = df.dropna()

print(df)
```

## 2.Text Preprocessing

```python
import re
import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer

nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('omw-1.4')

def preprocess_text(text):
    # Lowercasing
    
    text = text.lower()

    # Removing special characters and punctuation
    text = re.sub(r"[^a-zA-Z0-9]", " ", text)

    # Tokenization
    tokens = text.split()

    # Removing stop words
    stop_words = set(stopwords.words("english"))
    tokens = [word for word in tokens if word not in stop_words]

    # Removing numerical values
    tokens = [word for word in tokens if not word.isdigit()]

    # Removing extra whitespaces
    tokens = [word.strip() for word in tokens if word.strip()]

    # Joining tokens back to a single string
    processed_text = " ".join(tokens)
    return processed_text

# Applying text preprocessing to the 'Review' column of the DataFrame
df['Review'] = df['Review'].apply(preprocess_text)
```

## 3. Sentiment Analysis

### Using NLTK - Natural Language Toolkit
### The score for each row for positive, neutral, negative, and compounded

```python
import nltk
from nltk.sentiment import SentimentIntensityAnalyzer

nltk.download('vader_lexicon')
nltk.download('punkt')

def analyze_sentiment(text):
    sid = SentimentIntensityAnalyzer()
    sentiment_scores = sid.polarity_scores(text)
    sentiment = sentiment_scores['compound']
    
    print("sentiment scores:", sentiment_scores)
    
    tokens = nltk.word_tokenize(text)
    word_scores = []
    for token in tokens:
        word_score = sid.polarity_scores(token)
        word_scores.append((token, word_score))
    #print("word scores:", word_scores)
    
    if sentiment >= 0.05:
        return 'Positive'
    elif sentiment <= -0.05:
        return 'Negative'
    else:
        return 'Neutral'

    
# Applying sentiment analysis to the 'Review' column of the DataFrame
df['Sentiment_token'] = df['Review'].apply(analyze_sentiment)

# Print the sentiment for the first review
print(df)
```
