```python
import nltk
from nltk.sentiment import SentimentIntensityAnalyzer

nltk.download('vader_lexicon')
nltk.download('punkt')

def analyze_sentiment(text):
    sid = SentimentIntensityAnalyzer()
    sentiment_scores = sid.polarity_scores(text)
    sentiment = sentiment_scores['compound']

    tokens = nltk.word_tokenize(text)
    word_scores = []
    for token in tokens:
        word_score = sid.polarity_scores(token)
        word_scores.append((token, word_score))

    if sentiment >= 0.05:
        return 'Positive'
    elif sentiment <= -0.05:
        return 'Negative'
    else:
        return 'Neutral'

# Applying sentiment analysis to the content of each message in the JSON data
for message in json_data["messages"]:
    message_content = message["content"]
    sentiment = analyze_sentiment(message_content)
    message["sentiment"] = sentiment

# Printing the sentiment for each message
for message in json_data["messages"]:
    print("Message Content:", message["content"])
    print("Sentiment:", message["sentiment"])
    print("------")
```

```python
import plotly.express as px

# Calculate the count of each sentiment category
sentiment_counts = {}
for message in json_data["messages"]:
    sentiment = message["sentiment"]
    sentiment_counts[sentiment] = sentiment_counts.get(sentiment, 0) + 1

# Prepare data for the bar chart
sentiments = list(sentiment_counts.keys())
review_counts = list(sentiment_counts.values())

# Create the bar chart using Plotly Express
bar_chart = px.bar(x=sentiments, y=review_counts, title="Sentiment Analysis of Facebook Messages")
bar_chart.show()

```
